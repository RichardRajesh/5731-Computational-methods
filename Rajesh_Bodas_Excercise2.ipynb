## Question 2 (10 Points)
Write Python code to collect a dataset of 1000 samples related to the question discussed in Question 1.

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

Scrapping_url = 'https://www.imdb.com/list/ls098063263/?page={}'

movie_data = []

for page_num in range(1, 11):

    url_used = Scrapping_url.format(page_num)

    response = requests.get(url_used)

    beautysoup = BeautifulSoup(response.content, 'html.parser')

    movie_info = beautysoup.find_all('div', class_='lister-item mode-detail')

    for store in movie_info:

        name = store.h3.a.text.strip()

        director_tag = store.find('p', class_='text-muted text-small').find('a')
        director = director_tag.text.strip() if director_tag else 'Unknown'

        runtime = store.find('span', class_='runtime').text.strip(' min')

        rating = store.find('span', class_='ipl-rating-star__rating').text.strip()
        summary = store.find('p', class_='').text.strip()
        movie_data.append({
            'Movie Title': name,
            'Director': director,
            'Runtime': runtime,
            'IMDb Rating': rating,
            'Plot': summary
        })
movie_DF = pd.DataFrame(movie_data)
movie_DF.head(1000)
```

## Question 3 (10 Points)
Write Python code to collect 1000 articles from Google Scholar (https://scholar.google.com/), Microsoft Academic (https://academic.microsoft.com/home), or CiteSeerX (https://citeseerx.ist.psu.edu/index), or Semantic Scholar (https://www.semanticscholar.org/), or ACM Digital Libraries (https://dl.acm.org/) with the keyword "XYZ". The articles should be published in the last 10 years (2014-2024).

The following information from the article needs to be collected:

(1) Title of the article

(2) Venue/journal/conference being published

(3) Year

(4) Authors

(5) Abstract

```python
# write your answer here
import requests
from bs4 import BeautifulSoup


def extract_article_info(article):

    title = article.find('h3', class_='gs_rt').text.strip()


    metadata = article.find('div', class_='gs_a').text.split(' - ')
    year = metadata[0].strip()
    authors = metadata[1].strip()
    venue = metadata[-1].strip()


    abstract = article.find('div', class_='gs_rs').text.strip()

    return title, venue, year, authors, abstract


def webscrape_schol_articles(url):
    # Send an HTTP GET request to the provided URL
    response = requests.get(url)


    soup = BeautifulSoup(response.text, 'html.parser')


    articles = soup.find_all('div', class_='gs_ri')


    for article in articles:
        # Extract information from the article
        title, venue, year, authors, abstract = extract_article_info(article)


        print("Title:", title)
        print("Venue:", venue)
        print("Year:", year)
        print("Authors:", authors)
        print("Abstract:", abstract)
        print("\n---\n")


articles_per_page = 10
total_articles = 1000

for page in range(0, total_articles, articles_per_page):

    page_url = f"https://scholar.google.com/scholar?hl=en&as_sdt=0%2C5&q=information+retrieval&as_ylo=2013&as_yhi=2023&start={page}"

    webscrape_schol_articles(page_url)
```
